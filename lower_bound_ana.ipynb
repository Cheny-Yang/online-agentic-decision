# **1. Minimax Lower Bound**

### **1.1 Def: Worst-case regret**
 
define the **worst-case regret for policy $\pi$** as:

$$
R_n(\pi, \mathcal{E}) =  \sup_{v \in \mathcal{E}} R_n(\pi, v)
$$

### **1.2 Def: minimax regret**

define the **minimax regret over $\pi$ and $\mathcal{E}$** as:

$$
\begin{align}
R_n^*(\mathcal{E}) &:= \inf_{\pi \in \Pi} R_n(\pi, \mathcal{E}) \\
&= \inf_{\pi \in \Pi}\sup_{v \in \mathcal{E}} R_n(\pi, v)
\end{align}
$$

> **📌(1) Remark: $\inf_{\pi \in \Pi}\sup_{v \in \mathcal{E}} R_n(\pi, v)$ is a two-stage game**
>
> - **stage 1:** fix $\pi \in \Pi$, then a taughest instance $v^* \in \mathcal{E}$ for policy $\pi$ is chosen from the environment
>
> - **stage 2:** fix $v^*$, then choose the best $\pi^* \in \Pi$ to minimize the regret $R_n(x^*,v^*)$

> **❓Qustion: Is $\inf_{\pi \in \Pi}\sup_{v \in \mathcal{E}} R_n(\pi, v)$ really a robust optimization???**
>
> even if we understand $\inf_{\pi \in \Pi}\sup_{v \in \mathcal{E}} R_n(\pi, v)$ as a two-stage game, the choice of $v^*$ is dependent to the initial $\pi$, thus $R_n^*(\mathcal{E})$ might not be unique!?

> **📌(2) Remark: 🟥$R_n^*(\mathcal{E})$ is the best performance under the worst case**

### **1.3 Thr: in $\mathcal{E}^k$, each algorithm has its nemesis**

Let:

- $\mathcal{E}^k :=\{\text{all k-armed Gaussian bandits with} \, \sigma = 1 \,\text{and}\, \mu \in [0,1]^k \} $

then $\exists c >0$ s.t. for $\forall k \in \mathbb{N}_+$ and $\forall n \ge k$:

$$
R_n^*(\mathcal{E}^k) \ge c \sqrt{kn}
$$

it is equivalent to:

for $\forall \pi \in \Pi$, $\exists v^* \in \mathcal{E}^k$ s.t. 

$$
R_n(\pi, v^*) \ge c \sqrt{kn}
$$

$proof.$

the goal is to find a pair of instance $v$ and $v'$ satisfying the following condition:

- **Competition:** $\nexists (A_t)_{t=1}^n$ s.t.  $R_n((A_t)_{t=1}^n, v)$ and $R_n((A_t)_{t=1}^n, v')$ both achieves their minimum ❓is this correct?
- **Similarity:** the algorithm cannot statistically distinguish $v$ and $v'$ with reasonable statistical accuracy

> ❓ **how to mathematically describe the conditions above?** 

let:

- $\mu_i , \mu_i' \in [0,1]$
- $P_i = \mathcal{N}(\mu_i,1)$
- $P_i' = \mathcal{N}(\mu_i',1)$
- $v := (P_i)_{i=1}^k$
- $v' := (P_i')_{i=1}^k$



then claim that for $\forall \pi \in \Pi$, $\exists (\mu_i)_{i=1}^k, (\mu_i')_{i=1}^k$ s.t. 

$$
\max\{R_n(\pi,v), R_n(\pi,v')\} \ge c \sqrt{kn}
$$

the proof is as follows:

let $\Delta \in (0, 1/2]$ be a constant left to tuned later, set 

$$
\mu = (\Delta, 0, 0, \cdots, 0)
$$

then 

$$
R_n(\pi, v) = (n- \mathbb{E}[T_1(n)]) \Delta
$$

then we select the least-often chosen arm:

$$
i^* := \arg\min_{j > 1} \mathbb{E}[T_j(n)]
$$

then for $\forall j \in [k]/\{1\}$:

$$
\mathbb{E}[T_{i^*}(n)] \le \mathbb{E}[T_{j}(n)]
$$

$\Rightarrow$ $n = \mathbb{E}[T_1(n)] + \sum_{i=2}^k\mathbb{E}[T_i(n)] \ge (k-1) \cdot \mathbb{E}[T_{i^*}(n)]$ $\Rightarrow$ 
$$\mathbb{E}[T_{i^*}(n)] \le \frac{n}{k-1}$$
 

then we construct another $\mu' \in [0,1]^k$ s.t 

$$
\mu'_j =
\begin{cases}
\mu_j, & \text{if } j \neq i, \\
2\Delta, & \text{otherwise}.
\end{cases}
$$

i.e., we only change the least-often visit arm into $2 \Delta$ and keep other arm the same as $\mu$.

then the regret of this bandit should be

$$
R_n(\pi, v') = \Delta \mathbb{E}[T_1(n)] + \sum_{j \notin \{1,i^*\}} 2 \Delta \mathbb{E}[T_j(n)] \ge \Delta \mathbb{E}[T_1(n)]
$$

> **How to prove that $v$ and $v'$ are statistically impossible for $\forall \pi \in \Pi$ to distinguish** 
>

### **1.3.1 Lemma : Bretagnolle-Huber inequality** 

suppose $P$ and $Q$ are probability measures on the same measurable space $(\Omega, \mathcal{F})$, then for $\forall A \in \mathcal{F}$:

$$
P(A) + Q(A^c) \ge \frac{1}{2} \exp(- D(P,Q))
$$

> **📌Remark: which means that no matter how we design the hypothesis testing, the probability of failing the test will be no less than $\frac{1}{2} \exp(- D(P,Q))$**

then fix a policy $\pi \in \Pi$, then

- $v := (P_1, \cdots, P_k)$
- $v' := (P_1', \cdots, P_k')$
- $\Omega:= ([k] \times \mathcal{X})^n$
- $\mathcal{F}:= (\sigma([k]) \otimes \mathcal{B}(\mathcal{X}))^{\otimes n}$
- $\mathbb{P}_{v}:=$ the measure on $(\Omega, \mathcal{F})$ that the policy $\pi$ interact with $v$
- $\mathbb{P}_{v'}:=$ the measure on $(\Omega, \mathcal{F})$ that the policy $\pi$ interact with $v'$

### **1.3.2 Lemma : Divergence Decomposition**

$$
D(\mathbb{P}_{v}, \mathbb{P}_{v'}) = \sum_{i=1}^k \mathbb{E}_{v}[T_i(n)] D(P_i, P_i')
$$

$proof.$

suppose that

- $D(P_i, P_i') < \infty$ for $\forall i \in [k]$ $\Leftarrow$ $P_i \ll P'_i$

and define

- $\lambda := \sum_{i=1}^k P_i + P_i'$ $\Rightarrow$ $\lambda \gg P_i$ and $\lambda \gg P'_i$ for $\forall i \in [k]$

then we have 

$$
D(\mathbb{P}_{v}, \mathbb{P}_{v'}) = \mathbb{E}_v\left[\log \left(\frac{d \mathbb{P}_v }{d\mathbb{P}_{v'}}  \right)  \right]
$$

and let $\rho$ be the counting measure over $[k]$, then the Radon-Nikodym derivative of $\mathbb{P}_v$ w.r.t $(\rho \times \lambda)^n$ is 

$$
p_{v,\pi}(a_1, x_1, \cdots, a_n,x_n) = \prod_{t=1}^n \pi_{t}(a_t | a_1, x_1, \cdots, a_{t-1}, x_{t-1})p_{a_t}(x_t)
$$

and similarly, 

$$
p_{v',\pi}(a_1, x_1, \cdots, a_n,x_n) = \prod_{t=1}^n \pi_{t}(a_t | a_1, x_1, \cdots, a_{t-1}, x_{t-1})p'_{a_t}(x_t)
$$

then 

$$
\begin{align}
\log\left( \frac{d \mathbb{P}_{v}}{d\mathbb{P}_{v'}}  \right) 
&= \log\left( \frac{\prod_{t=1}^n p_{a_t}(x_t)}{\prod_{t=1}^n p'_{a_t}(x_t)}  \right) \\
&= \sum_{t=1}^n  \log\left(\frac{p_{a_t}(x_t)}{p'_{a_t}(x_t)}  \right)
\end{align}
$$

then

$$
\begin{align}
\mathbb{E}_v\left[\log \left(\frac{d \mathbb{P}_v }{d\mathbb{P}_{v'}}  \right)  \right]
&=  \mathbb{E}_v\left[ \sum_{t=1}^n  \log\left(\frac{p_{A_t}(X_t)}{p'_{A_t}(X_t)}  \right)  \right] \\
&= \sum_{(a_1, a_2, \cdots, a_n) \in [k]^n}\int_{\mathcal{X}^n} \left[\sum_{t=1}^n \log\left(\frac{p_{a_t}(x_t)}{p'_{a_t}(x_t)}\right)\right]\mathbb{P}_v(a_1,x_1, \cdots, a_n,x_n) dx_1dx_2 \cdots dx_n \\
&= \sum_{t=1}^n \sum_{a_t \in [k]}\int_{\mathcal{X}} \log\left(\frac{p_{a_t}(x_t)}{p'_{a_t}(x_t)}\right)\mathbb{P}_v(a_t,x_t | a_{t-1},x_{t-1},\cdots,a_1,x_1) dx_t \\
&= \sum_{t=1}^n \sum_{a_t \in [k]}\left[\int_{\mathcal{X}} \log\left(\frac{p_{a_t}(x_t)}{p'_{a_t}(x_t)}\right) \mathbb{P}_v(x_t | a_t, x_{t-1}, a_{t-1}, \cdots, x_1,a_1) dx_t \right] \mathbb{P}_v(a_t | x_{t-1}, a_{t-1}, \cdots, x_1,a_1)\\
&= \sum_{t=1}^n \sum_{a_t \in [k]}\left[ D(P_{a_t},P'_{a_t})\right]\mathbb{P}_v(a_t | x_{t-1}, a_{t-1}, \cdots, x_1,a_1) \\
&= \sum_{t=1}^n \mathbb{E}_v[D(P_{A_t},P'_{A_t})] \\
&= \sum_{t=1}^n \mathbb{E}_v\left[\sum_{i=1}^k \mathbb{I}\{A_t = i\}D(P_{A_t},P'_{A_t})\right]\\
&= \sum_{i=1}^k\mathbb{E}_v\left[\sum_{t=1}^n  \mathbb{I}\{A_t = i\}D(P_{A_t},P'_{A_t})\right]\\
&= \sum_{i=1}^k\mathbb{E}_v[T_i(n)]D(P_{i},P'_{i})
\end{align}
$$

complete the proof.$\blacksquare$

### **1.3.3 $\max\{R_n(\pi,v), R_n(\pi,v')\} \ge c \sqrt{kn}$**

By Lemma 1.3.2, we have

$$
\begin{align}
D(\mathbb{P}_v, \mathbb{P}_{v'}) &= \sum_{i=1}^k \mathbb{E}_{v}[T_i(n)] D(P_i, P_i') \\
&= \sum_{i=1}^k \mathbb{E}_{v}[T_i(n)] \frac{(\mu_i - \mu_i')^2}{2} \\
&= \mathbb{E}_{v}[T_{i^*}(n)] \frac{(\mu_{i^*} - \mu_{i^*}')^2}{2} \\
&= \mathbb{E}_{v}[T_{i^*}(n)] \frac{(2\Delta)^2}{2}
\end{align}
$$

since $\mathbb{E}[X] \ge a\mathbb{P}_v(X \ge a)$, then

$$
\begin{align}
R_n(\pi, v) &= (n- \mathbb{E}_v[T_1(n)]) \Delta \\
&= \mathbb{E}_v[n - T_1(n)]\Delta \\
& \ge \mathbb{P}_v(n - T_1(n) \ge n/2) \frac{\Delta n}{2} \\
&= \mathbb{P}_v(T_1(n) \le n/2) \frac{\Delta n}{2} \\
\end{align}
$$

$$
\begin{align}
R_n(\pi, v') &= \Delta \mathbb{E}[T_1(n)] + \sum_{j \notin \{1,i^*\}} 2 \Delta \mathbb{E}_{v'}[T_j(n)] \\
&\ge \Delta \mathbb{E}_{v'}[T_1(n)] \\
&\ge \frac{n\Delta}{2}\mathbb{P}_{v'}(T_1(n) \ge n/2)
\end{align}
$$

Then by Lemma 1.3.1, 

$$
\begin{align}
R_n(\pi,v) + R_n(\pi,v') &= \frac{\Delta n}{2}\left[\mathbb{P}_v(T_1(n) \le n/2)  + \mathbb{P}_{v'}(T_1(n) \ge n/2) \right] \\
&\ge  \frac{\Delta n}{4}\exp\left[-D(\mathbb{P}_v, \mathbb{P}_{v'})  \right] \\
&=  \frac{\Delta n}{4}\exp\left[-\mathbb{E}_{v}[T_{i^*}(n)] \frac{(2\Delta)^2}{2}\right] \\
&\ge  \frac{\Delta n}{4}\exp\left[- \frac{2n\Delta^2}{k-1}\right] \\
\end{align}
$$

then since $2\max(a,b) \ge a+b$, we have

$$
\begin{align}
\max\{R_n(\pi,v), R_n(\pi,v')\} 
&\ge \frac{1}{2}(R_n(\pi,v) + R_n(\pi,v'))\\
&\ge  \frac{\Delta n}{8}\exp\left[- \frac{2n\Delta^2}{k-1}\right] \\
\end{align}
$$

then let $\Delta = \sqrt{(k-1)/(4n)}$, since $n \ge k$, then $\Delta \le 1/2$, and consequently

$$
\frac{\Delta n}{8}\exp\left[- \frac{2n\Delta^2}{k-1}\right]  = \frac{\exp(-\frac{1}{2})}{16} \sqrt{(k-1)n}
$$

hence, complete the proof. $\blacksquare$

also, notice that in this case, we have:
$$
\begin{align}
D(\mathbb{P}_v, \mathbb{P}_{v'}) 
&= \mathbb{E}_{v}[T_{i^*}(n)] \frac{(2\Delta)^2}{2} \\
&\le \frac{n2\Delta^2}{k-1} \\
\end{align}
$$

so $D(\mathbb{P}_v, \mathbb{P}_{v'}) \rightarrow 0$ if $\Delta \rightarrow 0$

### **Pareto Optimality for multi-objective setting**

since the minimization to the  regret on a class of bandits $\mathcal{E}$ via selecting algorithm $\pi \in \Pi$ is a multi-objective task, i.e. , one $\pi$ will be good for some instance while bad on others, thus, we need some special way to analyse the performance of an algorithm, one way is the pareto optimality.

#### **Def: Pareto Optimality**

a policy $\pi$ is Pareto optimal if $\nexists \pi'$ s.t. for $\forall v \in \mathcal{E}$:

$$
R_n(\pi',v) \le R_n(\pi,v) 
$$

despite that there exists at least one $v \in \mathcal{E}$ s.t. $R_n(\pi',v) < R_n(\pi,v)$

# **2. Instance-Dependent Lower Bounds**

## **2.1 Asymptotic Bounds** 

### **Def: consistent policy**

call a policy consistent over a class of bandits $\mathcal{E}$ if for $\forall v \in \mathcal{E}$ and $p > 0$:

$$
\lim\limits_{n \rightarrow \infty} \frac{R_n(\pi,v)}{n^p} = 0
$$

and denote the class of consistent policies over $\mathcal{E}$ as $\Pi_{cons}(\mathcal{E})$

> **📌Remark:this criteria doesn't measure the efficiency**
