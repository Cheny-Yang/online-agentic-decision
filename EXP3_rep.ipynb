## **1.1 Environment**

- Oblivious Adversary


| Symbol | Definition |
|-------|------------|
| $k \in \mathbb{N}_+$  ($k < \infty$)| total number of arms  |
| $n \in \mathbb{N}_+$  | time horizon   |
| $\mathcal{X}(k,n):= [0,1]^{k \times n}$  | space of all feasible oblivious adversarial rewards   |

> **đź“Ś Remark: representation of one environment instance**
>
> any instance of the adversarial bandit environment with $k$ arms and $n$ time hirizon can be uniquely represented by a $k \times n$ matrix $x \in \mathcal{X}(k,n)$, where the $t$-th column represent the reward vector for step $t$.
>
> 

## **1.2 Decision Variables**

| Symbol | Definition |
|-------|------------|
| $A_t \in [k]$ | The arm chosed at step $t$|

## **1.3 Regret** 

$$
R_n(\pi, x) := \max\limits_{i \in [k]}\sum_{t=1}^n x_{i,t} - \mathbb{E}\left[\sum_{t=1}^n x_{A_t,t}  \right]
$$

> **đź“ŚRemark: The determination of "the best arm" is the same as that of the stochastic bandit**
>
> In this benchmark, the best ocasion is defined by choosing **the one arm** that maximize the **total reward**, i.e. $i^* := \arg\max\limits_{i\in [k]} \sum_{t=1}^n x_{i,t}$, this definition of regret retrict that the best learner for comparison can only stick to one arm along the horizon, however, $\max\limits_{i\in [k]} \sum_{t=1}^n x_{t,i} \le \sum_{t=1}^n \max\limits_{i \in [k]} x_{t,i}$.

> **đź“ŚRemark: No randomness comes from the environment**
> 
> Since the adversarial reward is oblivious, then $(x_t)_{t=1}^n$ is a sequence of constant real vectors, hence there is no randomness from the environment.


the worst-case regret over all environments is:

$$
R^*_n(\pi) := \sup_{x \in \mathcal{X}(k,n)} R_n(\pi, x)
$$

### **1.3.1 Thr: No deterministic policy can achieve a sublinear regret in this adversarial bandit game**


For any deterministic policy $\pi$, there exists an environment $x \in \mathcal{X}(k,n)$ s.t. 

$$
R_n(\pi, x) \ge n (1 - \frac{1}{k})
$$

> **Def: deterministic policy**
>
> Let $\mathcal{H}_t := (a_1, x_1, \cdots, a_{t-1}, x_{t-1})$,if a policy $\pi$ is deterministic, then for $\forall t \in [n]$: $A_t  = \pi_t(\mathcal{H}_t)$. i.e., $\pi_t$ is a well-defined map from the set of $\mathcal{H}_t$ to the set of arms $[k]$, the same $\mathcal{H}_t$ will always lead to the same $A_t$, no surprise. In another word, if $\pi$ is deterministic, then given a fix instance of environment $x$, the sequence of action $(A_1, \cdots, A_n)$ is known. 

$proof.$

For an arbitrary deterministic policy $\pi$, suppose the adversary has gotten full information about this policy, i.e. the map $\pi: \Omega \rightarrow [k]$ is unrevealed for the adversary, then the adversary can set the $x \in \mathcal{X}$ s.t. 

$\exists x \in \mathcal{X}$ s.t 

$$
x_{i,t}  = \mathbb{I}\{ i = a^* \}
$$

where $a^*$ is the least frequently chosen arm. Then since $\sum_{i=1}^k \mathbb{E}[T_i(n)] = n$ and $\mathbb{E}[T_{a^*}(n)] \le \mathbb{E}[T_i(n)]$ ($i \in [k]$), then $k\cdot \mathbb{E}[T_{a^*}] \le \sum_{i=1}^k \mathbb{E}[T_i(n)] = n$ $\Rightarrow$ $\mathbb{E}[T_{a^*}(n)] \le \frac{n}{k}$, thus 

$$
\begin{align}
R_n &= \max\limits_{i \in [k]}\sum_{t=1}^n x_{i,t} - \mathbb{E}\left[\sum_{t=1}^n x_{A_t,t}  \right] \\
&= n - \mathbb{E}\left[\sum_{t=1}^n x_{A_t,t}  \right] \\
&\ge n - \frac{n}{k}
\end{align}
$$

complete the proof. $\blacksquare$.




> **đź“Ś Discusion: another way to construct adversary reward**
>
> 
> Learner choose an arm $A_t \in [k]$, which may not be able to known for the adversary
> 
> For $t=2, \cdots, n$:
> 
> $~~~~~~$ Adversary has known $A_t = \pi_t(\mathcal{H}_{t})$
>
> $~~~~~~$ Adversary set $x_{A_t, t} = 0$, $x_{i,t} = 1$ for $i \in [k]/A_t$ $\Rightarrow$ the regret at step $t$ is 1
